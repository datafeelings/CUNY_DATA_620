**Solution**

```python
import networkx as nx

from bokeh.io import show, output_file
from bokeh.models import Plot, Range1d, MultiLine, Circle, HoverTool, TapTool, BoxSelectTool
from bokeh.models.graphs import from_networkx, NodesAndLinkedEdges, EdgesAndLinkedNodes
from bokeh.palettes import Spectral4
```

```python
G = nx.Graph()
```

```python
names = ["Andre","Beverly","Carol","Diane","Ed","Fernando","Garth","Heather","Ike","Jane"]
G.add_nodes_from(names)
```

```python
edges = [("Andre","Carol"),("Andre","Diane"),("Andre","Beverly"),("Andre","Fernando"),
         ("Beverly","Ed"),("Beverly","Diane"),("Beverly","Garth"),("Carol","Diane"),
        ("Carol","Fernando"),("Diane","Ed"),("Diane","Fernando"),("Diane","Garth"),
        ("Garth","Ed"),("Garth","Fernando"),("Garth","Heather"),("Fernando","Heather"),
        ("Heather","Ike"),("Ike","Jane")]
G.add_edges_from(edges)
```

I did not consider the fact that names are actually just attributes of the nodes. So I have to replace the node ids with integers and reassign the name attributes:

```python
G=nx.convert_node_labels_to_integers(G)

for item in G.nodes():
    G.add_node(item, name=names[item])

G.nodes(data=True)  
```

```python
plot = Plot(plot_width=700, plot_height=600,
            x_range=Range1d(-1.1,1.1), y_range=Range1d(-1.1,1.1))
plot.title.text = "Krackhardt Kite Graph"

plot.add_tools(HoverTool(tooltips=None), TapTool(), BoxSelectTool())

graph_renderer = from_networkx(G, nx.spring_layout, scale=1.3, center=(0,0))

graph_renderer.node_renderer.glyph = Circle(size=15, fill_color=Spectral4[0])
graph_renderer.node_renderer.selection_glyph = Circle(size=15, fill_color=Spectral4[2])
graph_renderer.node_renderer.hover_glyph = Circle(size=24, fill_color=Spectral4[1])

graph_renderer.edge_renderer.glyph = MultiLine(line_color="#CCCCCC", line_alpha=0.8, line_width=5)
graph_renderer.edge_renderer.selection_glyph = MultiLine(line_color=Spectral4[2], line_width=5)
graph_renderer.edge_renderer.hover_glyph = MultiLine(line_color=Spectral4[1], line_width=5)


graph_renderer.selection_policy = NodesAndLinkedEdges()
graph_renderer.inspection_policy = EdgesAndLinkedNodes()

plot.renderers.append(graph_renderer)

output_notebook()
```

```python
show(plot)
```
